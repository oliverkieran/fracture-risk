```python
data = pd.read_csv("/mnt/project/data/processed/icd10_codes.csv")
data
```

```python
FX_CODES = {
    "vertebral": ["S12.0", "S12.1", "S12.2", "S12.7", "S22.0", "S22.1", "S22.4", "S32.0", "S32.1", "S32.7", "M48.44", "M48.45", "M48.46", "M48.47", "M48.48", "M48.49"],
    "hip": ["S72.0", "S72.1", "S72.2"],
    "wrist": ["S62.0", "S62.1", "S62.2", "S62.3", "S62.4"],
}
FX_CODES["any"] = FX_CODES["vertebral"] + FX_CODES["hip"] + FX_CODES["wrist"] + ["M80", "S42", "S52", "S72", "S82", "S92.0", "S92.1", "S92.2", "S92.3", "S92.7", "S92.9"]
```

```python
# Define date and value columns
values = data.loc[:,"p41270_a0":"p41270_a118"]
dates = data.loc[:,"p41280_a0":"p41280_a118"]

# We want to extract only the dates where the code has been found
values = pd.concat(
    [
        data[value_column].str.startswith(tuple(FX_CODES["any"]))
        for value_column in values.columns
    ],
    axis=1
)
dates = pd.concat(
    [
        pd.to_datetime(data[date_column]).where(values[value_column], pd.NaT)
        for value_column, date_column in zip_longest(values.columns, dates.columns)
    ],
    axis=1
)
values
```

```python
non_nans = data.loc[:,"p41270_a0":"p41270_a118"][data.loc[:,"p41270_a0":"p41270_a118"][values.fillna(False)].notna().any(axis=1)].copy()
non_nans
```

```python
# Check if there are any icd-10 codes that were recorded multiple times per patient.
icd_codes = non_nans.notna().sum(axis=1)
(non_nans.apply(lambda x : x.value_counts() , axis = 1) > 1).sum(axis=1).sum()
```
